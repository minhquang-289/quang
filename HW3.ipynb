```python
import sys

def main():
    data = sys.stdin.read().strip().split()
    if not data:
        return
    it = iter(data)
    t = int(next(it))
    INF = 10**30

    out_lines = []

    for _ in range(t):
        n = int(next(it))
        a = [int(next(it)) for _ in range(n)]

        # Prefix sums S[0..n]
        S = [0] * (n + 1)
        for i in range(1, n + 1):
            S[i] = S[i - 1] + a[i - 1]

        # Trường hợp không “tận dụng” đảo đoạn (tức chọn prefix tốt nhất gốc)
        base = max(S)  # đã gồm S[0] = 0

        # sufMax[i] = max_{k >= i} S[k]
        sufMax = [-INF] * (n + 2)
        sufMax[n] = S[n]
        for i in range(n - 1, 0, -1):
            sufMax[i] = max(S[i], sufMax[i + 1])

        # Tính hiệu ứng đảo đoạn qua từng j
        max_extra = -INF
        prefMax = S[0]  # max S[0..j-1]
        for j in range(1, n):
            bestRight = sufMax[j + 1]  # max S[r] với r > j
            cand = prefMax + bestRight - S[j]
            if cand > max_extra:
                max_extra = cand

            if S[j] > prefMax:
                prefMax = S[j]

        ans = max(0, base, max_extra)
        out_lines.append(str(ans))

    sys.stdout.write("\n".join(out_lines))


if __name__ == "__main__":
    main()
```
