```python
#W7A22
#Cho N tên sinh viên gồm cả họ và tên,
#tìm xem tên nào hay gặp nhất. Nếu có
#nhiều tên cùng số lần gặp thì lấy tên có
#thứ tự từ điển nhỏ nhất

n=int(input())
cnt={}
for i in range(n):
    a=input().strip()
    name=""
    for i in range(len(a)-1,-1,-1):
        if a[i] == ' ':
            break
        name=a[i] + name
    if name in cnt:
        cnt[name]+=1
    else:
        cnt[name] = 1
max_fre=max(cnt.values())
res=""
for name in cnt:
    if cnt[name]==max_fre:
        if res=="" or name < res:
            res=name
print(res)



```
