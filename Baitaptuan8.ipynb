```python
#W7A22
#Cho N tên sinh viên gồm cả họ và tên,
#tìm xem tên nào hay gặp nhất. Nếu có
#nhiều tên cùng số lần gặp thì lấy tên có
#thứ tự từ điển nhỏ nhất

n=int(input())
cnt={}
for i in range(n):
    a=input().strip()
    name=""
    for i in range(len(a)-1,-1,-1):
        if a[i] == ' ':
            break
        name=a[i] + name
    if name in cnt:
        cnt[name]+=1
    else:
        cnt[name] = 1
max_fre=max(cnt.values())
res=""
for name in cnt:
    if cnt[name]==max_fre:
        if res=="" or name < res:
            res=name
print(res)



```

```python
#W7A27
def is_valid(s):
    stack=[]
    for ch in s:
        if ch =='(' or ch == '[' or ch == '{':
            stack.append(ch)
        else:
            if not stack:
                return False
            top = stack.pop()
            if ch ==')' and top !='(':
                return False
            if ch == ']' and top !='[':
                return False
            if ch == '}' and top != '{':
                return False
    return len(stack) == 0

if __name__=='__main__':
    s=input()
    print(is_valid(s))
```

```python
#W7A28
#Bà ̀ i toán Tìm dãy các hợp số dài nhất liên 
#tiếp không chứa số nguyên nào <= N (N 
#là số nguyên dương cho trước). Nếu có 
#nhiều dãy cùng độ dài thì lấy dãy nhỏ 
#nhất (gặp đầu tiên từ trái qua)
import math

def is_composite(n):
    if n <= 3:
        return False 
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return True
    return False


def longest_composite_sequence(N):
    max_len = 0
    max_start = -1

    current_len = 0
    current_start = -1

    for i in range(1, N + 1):
        if is_composite(i):
            if current_len == 0:
                current_start = i
            current_len += 1
        else:
            if current_len > max_len:
                max_len = current_len
                max_start = current_start
            current_len = 0

    if current_len > max_len:
        max_len = current_len
        max_start = current_start

    return max_start, max_len


if __name__ == '__main__':
    N = int(input())
    start, length = longest_composite_sequence(N)
    if length != 0 :
        for i in range(start, start + length):
            print(i, end=" ")
```

```python
#W7A31
#Đổi chỗ đúng 2 chữ số để được số nhỏ
#nhất có thể
s = list(input().strip())
n = len(s)

for i in range(n):
    min_digit = s[i]
    pos = -1

    for j in range(n-1, i, -1):
        if s[j] < min_digit:
            min_digit = s[j]
            pos = j

    if pos != -1:
        s[i], s[pos] = s[pos], s[i]
        break

print("".join(s))
```
