```python
#W9A1
def merge_sorted(a, b):
    i = j = 0
    merged = []

    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            merged.append(a[i])
            i += 1
        else:
            merged.append(b[j])
            j += 1

    merged.extend(a[i:])
    merged.extend(b[j:])
    return merged


print(merge_sorted([1, 3, 5, 9], [2, 3, 6, 10]))
```

```python
#W9A2
n, m, k = map(int, input().split())

a = []
for _ in range(n):
    a.append(list(map(int, input().split())))

ks = n * m - k + 1

lo = a[0][0]
hi = a[n - 1][m - 1]

while lo < hi:
    mid = (lo + hi) // 2
    cnt = 0
    i = n - 1
    j = 0
    while i >= 0 and j < m:
        if a[i][j] <= mid:
            cnt += i + 1
            j += 1
        else:
            i -= 1
    if cnt < ks:
        lo = mid + 1
    else:
        hi = mid

print(lo)
```

```python
#W9A3
s = list(input())
n = len(s)

i = 0
while i < n:
    j = i + 1
    pos = i
    while j < n:
        if s[j] < s[pos]:
            pos = j
        j += 1
    if pos != i:
        s[i], s[pos] = s[pos], s[i]
        break
    i += 1

print("".join(s))
```

```python
#W9A4
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    n = len(a)
    m = len(b)

    dp = [False] * (m + 1)
    dp[0] = True

    i = 0
    while i < n:
        ndp = [False] * (m + 1)
        ch = a[i]
        if 'a' <= ch <= 'z':
            up = chr(ord(ch) - 32)
            j = 0
            while j <= m:
                if dp[j]:
                    ndp[j] = True
                    if j < m and up == b[j]:
                        ndp[j + 1] = True
                j += 1
        else:
            j = 0
            while j < m:
                if dp[j] and ch == b[j]:
                    ndp[j + 1] = True
                j += 1
        dp = ndp
        i += 1

    print("YES" if dp[m] else "NO")
```

```python
#W9A5
def is_subarray(list1, list2):
    m, n = len(list1), len(list2)
    if m == 0:
        return True   
    if m > n:
        return False
    for i in range(n - m + 1):
        match = True
        for j in range(m):
            if list2[i + j] != list1[j]:
                match = False
                break
        if match:
            return True

    return False

print(is_subarray([2, 3, 4], [1, 2, 3, 4, 5]))
print(is_subarray([2, 4], [1, 2, 3, 4, 5])) 
```

```python
#W9A6
n = int(input())
a = list(map(int, input().split()))

i = n - 2
while i >= 0 and a[i] >= a[i + 1]:
    i -= 1

if i >= 0:
    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]

l = i + 1
r = n - 1
while l < r:
    a[l], a[r] = a[r], a[l]
    l += 1
    r -= 1

print(*a)
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    i = 0
    while i < n:
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i - 1] == 0)
            right = (i == n - 1) or (flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i] = 1
                k -= 1
                if k == 0:
                    return True
                i += 1
        i += 1
    return k <= 0
```

```python
#W9A8
def is_path_crossing(moves):
    x = 0
    y = 0
    seen = {(0, 0)}
    for ch in moves:
        if ch == 'R':
            x += 1
        elif ch == 'L':
            x -= 1
        elif ch == 'U':
            y += 1
        else:
            y -= 1
        if (x, y) in seen:
            return True
        seen.add((x, y))
    return False
```

```python
def Median(nums):
    a, b, c, d, e = nums

    if a > b:
        a, b = b, a
    if c > d:
        c, d = d, c

    if a > c:
        a, c = c, a
        b, d = d, b

    if b > e:
        b, e = e, b
    if b > c:
        b, c = c, b

    if b > d:
        b, d = d, b

    return b
```

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    j = 0
    res = 0
    m = len(heaters)

    for h in houses:
        while j + 1 < m and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
            j += 1
        d = heaters[j] - h
        if d < 0:
            d = -d
        if d > res:
            res = d
    return res
```
