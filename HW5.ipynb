```python
#W5A1
#Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two(
#a, b) -> int:
def Max_Of_Two(a,b):
    if a >b:
        return a
    else:
        return b
if __name__=='__main__':
    a,b=map(int,input().split())
    print (Max_Of_Two(a,b))
    

```

```python
#W5A2
# Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]: 
def swap(a,b):
    a ,b=b,a
    return a,b
if __name__=='__main__':
    a,b=map(int,input().split())
    print(swap(a,b))
```

```python
#W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
def check(n):
    for i in range(2,n-1):
        if n % i == 0:
            return False
        else:
            return True
if __name__=='__main__':
    n=int(input())
    print(check(n))
    
```

```python
#W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
# SO hoan hao la so tong cac uoc bang chinh nó
def check(n):
    sum =0
    for i in range(1,n-1):
        if n % i ==0:
            sum=sum+i
    if sum ==n:
        return True
    else:
        return False
if __name__=='__main__':
    n = int(input())
    print(check(n))
```

```python
#W5A5:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
#phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1
def lst(A,k):
    
    for i in range(len(A)):
        if A[i] == k:
            
            return i
    return -1
if __name__=='__main__':
    A = [3,4,5,1]
    k= int(input())
    print(lst(A,k))
```
