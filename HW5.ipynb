```python
#W6A16
#Nhâ ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
#trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải
n, m = map(int, input().split())

a = []
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)

for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}", end="")
    print()

```

```python
#W6A19
#Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
def max(A):
    m=A[0]
    for i in A:
        if i>m:
            m=i
    return i
if __name__=='__main__':
    A=list(map(int,input().split()))
    print(max(A))
```

```python
#W6A20
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
#trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
#-1
def so(A,k):
    for i in A:
        if k==i:
            return i
    else:
        return -1
if __name__=='__main__':
    A=list(map(int,input().split()))
    k=int(input())
    print(so(A,k))


```

```python
#W6A22
#Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
#trong list
def max(A):
    m=A[0]
    for i in range(1,len(A)):
        if A[i]>m:
            m=A[i]
    return m
if __name__=='__main__':
    A=list(map(int,input().split()))
    print(max(A))
```

```python
#W6A23
#VIết hàm xóa đi các số <x có trong mảng
def dl(A):
    for i in A:
        if i >=x:
            B.append(i)
    return B
if __name__=='__main__':
    A=list(map(int,input().split()))
    B=[]
    x=int(input())
    print(dl(A))

```

```python
#W6A24
#Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
#trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9
def day(A):
    cout=0
    for i in A:
        if i==0:
            cout+=1
        else:
            B.append(i)
    for k in range (cout):
        B.append(0)
    return B
if __name__=='__main__':
    A=list(map(int,input().split()))
    B=[]
    print(day(A))
```

```python
#W6A26
#Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
#phần tử đầu và cuối
a = list(map(int, input().split()))
b=[]
for i in range(1,len(a)-1):
    if a[i]>a[i+1] and a[i]>a[i-1]:
        b.append(a[i])
print(sum(b))

```

```python
#W6A18
#Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
#cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
#k (hàng k)
n,m,k=map(int,input().split())
a=[]
for i in range(n):
    b=list(map(int,input().split()))
    a.append(b)
tong_cot=0
tong_hang=0
for i in range(n):
    tong_cot+=a[i][k]
for j in range(m):
    tong_hang+=a[k][j]
print(tong_cot,tong_hang)
```
