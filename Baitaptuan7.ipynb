```python
#W7A1
#Viết hàm tìm kiếm nhị phân
#trên một danh sách đã
#được sắp xếp tăng dần.
#Trả về chỉ số của phần tử
#hoặc -1 nếu không tìm
#thấy. mẫu hàm: def
#binary_search(arr, left,
#right, target)
def binary_search(a,b):
    left,right=0,len(a)-1
    
    
    while left <= right: 
        mid = (left + right) // 2
        if a[mid] == b:
            return mid
        elif a[mid] < b:
            left = mid + 1 
        else:
            right = mid - 1  

    return -1  


if __name__=='__main__':
    a=list(map(int,input().split()))
    a.sort()
    b=int(input())
    print(binary_search(a,b))

```

```python
#W7A3
#Viết hàm thực hiện thuật
#toán bubble sort, Selection
#Sort để sắp xếp danh sách
#theo thứ tự tăng dần. Sắp
#xếp nổi bọt (Bubble Sort);
#Sắp xếp chọn (Selection
#Sort)
#Bubble sort 
def bubble_sort(a):
    n=len(a)
    for i in range(n):
        swapped=False
        for j in range(0,n-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
                swapped=True
        if not swapped:
            break
    return a


if __name__=='__main__':
    a=list(map(int,input().split()))
    print(bubble_sort(a))





```

```python
#Selection sort
def selection_sort(a):
    n=len(a)
    for i in range(n):
        min_idx=i
        for j in range(i+1,n):
            if a[j]<a[min_idx]:
                min_idx=j
        a[i],a[min_idx]=a[min_idx],a[i]
    return a

if __name__=='__main__':
    a=list(map(int,input().split()))
    print(selection_sort(a))
```

```python
#W7A7
#Cho một danh sách các số
#nguyên nums, hãy tìm độ
#dài của dãy con tăng
#nghiêm ngặt dài nhất
#(không nhất thiết phải liền kề)

a=eval(input())
c=[]
for x in a:
    l,r=0,len(c)
    while l < r:
        m=(l+r)//2
        if c[m] < x :
            l=m+1
        else:
            r=m
    if l == len(c):
        c.append(x)
    else:
        c[l]=x
print(len(c))
```

```python
#W7A8
#Bạn được cung cấp: một
#danh sách các khoảng thời
#gian interval = [[l1, r1], [l2,
#r2], ...] và một danh sách
#các truy vấn queries = [q1,
#q2, ...]
#Với mỗi truy vấn q, hãy tìm
#độ dài nhỏ nhất của
#khoảng thời gian [li, ri] sao
#cho li ≤ q ≤ ri.
#Nếu không tồn tại khoảng
#thời gian nào như vậy, hãy
#trả về -1 cho truy vấn đó.
#solve in < O(m.n), hint:
#binary search in the
#queries list

def solve(intervals,queries):
    intervals.sort(key=lambda x: x[1] - x[0])
    for q in queries:
        ans=-1
        for l,r in intervals:
            if l<=q<=r:
                ans=r-l+1
                break
        b.append(ans)
    return b


if __name__=='__main__':
    intervals=eval(input())
    queries=eval(input())
    b=[]
    print(solve(intervals,queries))
    
```

```python
#W7A10
#Cho một danh sách các số
#nguyên nums và một số
#nguyên m, hãy tìm tổng
#mảng con lớn nhất theo
#modulo m.
#Tức là, với bất kỳ mảng
#con nums[i:j] nào, hãy tính
#sum(nums[i:j]) % m và trả
#về kết quả lớn nhất có thể.
#Gợi ý:
#Duy trì tổng tiền tố (prefix
#sum) mod m và sử dụng
#tìm kiếm nhị phân để tìm
#tổng tiền tố nhỏ nhất chỉ
#lớn hơn tổng hiện tại một
#chút (để tối đa hóa hiệu số
#bao quanh).

a=eval(input())
m=int(input())
n=len(a)
prefix=0
b=[]
res=0
for i in range(n):
    prefix=(prefix+a[i]) % m
    if prefix>res:
        res=prefix
    for j in b:
        val=(prefix-j+m)%m
        if val > res:
            res=val
    b.append(prefix)
print(res)


    




```

```python
#W7A13
#Dãy hài hòa là một dãy
#thỏa mãn điều kiện giá trị
#lớn nhất và giá trị nhỏ nhất
#của dãy khác nhau đúng
#bằng \( 1 \).
#Viết hàm trả về độ dài của
#dãy hài hòa con dài nhất có
#thể có trong mảng giá trị
#nguyên \( arr \) có \( n \)
#phần tử
#Ví dụ:
#Input: 1 3 2 2 5 2 3 7
#Output: 5 - Dãy hài hòa con
#dài nhất là [3,2,2,2,3]

def longest_harmonious_subsequence(arr):
    b={}
    for i in arr:
        if i in b:
            b[i] += 1
        else:
            b[i] = 1
    max_len=0
    for i in b:
        if i+1 in b:
            lenght=b[i] + b[i+1]
            if lenght > max_len:
                max_len=lenght

    return max_len

if __name__=='__main__':
    arr=map(int,input().split())
    print(longest_harmonious_subsequence(arr))
```
