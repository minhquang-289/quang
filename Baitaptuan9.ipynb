```python
# [Cylinder]
# Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và
# chiều cao (kiểu số thực) với hai phương thức:tính
# diện tích bề mặt (diện tích toàn phần) và thể tích
# của hình trụ
# Đầu vào
# Đâu vào từ bàn phím gồm một dòng duy nhất
# chứa bán kính và chiều cao của hình trụ, cách
# nhau bởi một dấu cách.
# Đầu ra
# In ra màn hình diện tích bề mặt và thể tích của
# hình trụ (sử dụng lại những hàm vừa viết, π=3.
# 14).
# (Kết quả làm tròn đến 2 chữ số sau dấu phẩy).
class Cylinder:
    pi=3.14
    
    def __init__(self,r,h):
        self.x=r
        self.y=h
    def S(self):
        Sbm=2*Cylinder.pi*self.x*self.y+Cylinder.pi*self.x**2
        return f"{Sbm:.2f}"
    def V(self):
        V_=Cylinder.pi*(self.x)**2*self.y
        return f"{V_:.2f}"
    def __str__(self):
        return f"{self.S()} {self.V()}"
a,b=map(float,input().split())
hinhtru=Cylinder(a,b)
print(hinhtru)
```

```python
# [Date]
# Xây dựng lớp Date biểu diễn ngày tháng năm với
# hai phương thức kiểm tra xem một ngày tháng
# nào đó có hợp lệ hay không và phương thức tính
# ngày kế tiếp của ngày tháng năm nào đó nếu nó là
# hợp lệ
# Nhập vào một ngày tháng năm, (ngày tháng năm
# biểu diễn dưới dạng dd/mm/yyyy). Nếu ngày
# tháng năm không tồn tại, in ra dòng chữ INVALID.
# Nếu là hợp lệ in ra ngày kế tiếp của nó.
# Đầu vào
# Đầu vào từ bàn phim chứa dữ liệu ngày, tháng,
# năm dưới dạng dd/mm/yyyy .
# Đầu ra
# In ra màn hình ngày kế tiếp của ngày vừa nhập
# dưới dạng dd/mm/yyyy
# . Nếu ngày tháng không tồn tại, in ra INVALID
class date:
    def __init__(self,day,month,year):
        self.d=day
        self.m=month
        self.y=year
    def nam_nhuan(self):
        return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
    def ngay_cuoi_trong_thang(self):
        c=[1,3,5,7,8,10,12]
        d=[4,6,9,11]
        if self.m in c:
            return 31
        elif self.m in d:
            return 30
        elif self.m==2:
            if self.nam_nhuan():
                return 29
            else:
                return 28
    def ngay_tiep_theo(self):
        if self.d == self.ngay_cuoi_trong_thang():
            self.d=1
            self.m+=1
            if self.m>12:
                self.m=1
                self.y+=1
        else:
            self.d+=1
    def __str__(self):
        return f"{self.d:02d} / {self.m:02d} / {self.y}"
d,m,y=map(int,input().split())
ngay_thang_nam=date(d,m,y)
print("ngay hien tai: ",ngay_thang_nam)
ngay_thang_nam.ngay_tiep_theo()
print("ngay tiep theo : ",ngay_thang_nam)v
```

```python
# Xây dựng lớp điểm với hai thuộc tính x, y cùng
# các phương thức kiểm tra xem có phải gốc tọa dộ
# không, có nằm trên trục tung không, có nằm trên
# trục hoành không, tính khoảng cách từ điểm đó tới
# một điểm khác
# Đầu vào:
# Nhập vào tọa độ một điểm M bất kỳ
# Đầu ra:
# Thông báo M có phải gốc tọa độ không hoặc có
# nằm trên trục tung hoặc trục hoành không, khoảng
# cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ
# số phần thập phân)
import math
class truc_toa_do:
    def __init__(self,a,b):
        self.x=a
        self.y=b
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def nam_tren_truc_tung(self):
        return self.x==0 and self.y!=0
    def nam_tren_truc_hoanh(self):
        return self.y==0 and self.x!=0
    def khong_nam_tren_truc_nao(self):
        return self.y!=0 and self.x!=0
    def khoang_cach_den_o(self):
        return math.sqrt((self.x)**2+(self.y)**2)
    def __str__(self):
        if self.goc_toa_do():
            vi_tri="M là gốc tọa độ"
        elif self.nam_tren_truc_hoanh():
            vi_tri="M nằm trên trục hoành"
        elif self.nam_tren_truc_tung():
            vi_tri="M nằm trên trục tung"
        elif self.khong_nam_tren_truc_nao():
            vi_tri="M không nằm trên trục nào"
        return f"{vi_tri}. Khoảng cách đến gốc tạo độ là {self.khoang_cach_den_o():.2f}"
a,b=map(int,input().split())
M=truc_toa_do(a,b)
print(M)
```

```python
# Xây dựng lớp Calculator (máy tính bỏ túi) với các
# phương thức để thực hiện các phép toán 1. add
# (cộng), 2. subtract(trừ), 3. product (nhân) và 4.
# divide(chia), 5. power(hàm mũ), 6. mod (lấy phần
# dư) và phương thức 7. set_numbers(cập nhật 2
# số mới) nếu người dùng muốn tính toán trên 2 số
# khác.
# Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính,
# Mỗi phương thức thực hiện một phép tính và trả
# về kết quả tương ứng
# Tạo menu và cho người dùng tính toán tới khi gõ
# "yes", "YES" hoặc "Yes" cho câu hỏi "Bạn có
# muốn thoát?" mới dừng
class Calculator:
    def __init__(self,a,b):
        self.x=a
        self.y=b
    def add(self):
        return self.x + self.y
    def subtract(self):
        return self.x - self.y
    def product(self):
        return self.x * self.y
    def divide(self):
        if self.y==0:
            return "khong the chia cho 0"
        return self.x / self.y
    def power(self):
        return self.x**self.y
    def mod(self):
        return self.x % self.y
    def set_number(self,e,f):
        self.x=e
        self.y=f
        return self.x and self.y
    def dung_chuong_trinh(self):
        while self.x=="yes" or self.y=="yes":
            break
    def __str__(self):
        return (f"Phep cong: {self.add()}\n"
            f"Phep tru: {self.subtract()}\n"
            f"Phep nhan: {self.product()}\n"
            f"Phep chia: {self.divide()}\n")
a,b=map(float,input().split())
M=Calculator(a,b)
flag=False
while True:
    if not flag:
        print(M)
    n=input()
    if n.lower()=="yes":
        print("bạn đã chọn dừng chương trình")
        break
    doi_so=input()
    if doi_so=="ok":
        e,f=map(float,input().split())
        M.set_number(e,f)
        print("đổi 2 số ban đầu thành: ",e ,"và" ,f)
        flag=True
```

```python
# [ShoppingCart]
# Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý
# các sản phẩm của khách. Dùng danh sách để lưu
# các sản phẩm, mỗi sản phẩm là một từ điển có
# name (tên) và price(giá).
# Với các phương thức sau:
# 1. Khởi tạo với danh sách rỗng
# 2. Thêm vào giỏ hàng một sản phẩm mới
# 3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên
# để xóa)
# 4. Kiểm tra xem giỏ hàng có rỗng không
# 5. Tính tổng tiền giỏ hàng
# 6. Hiển thị toàn bộ sản phẩm trong giỏ
# 7. Xóa toàn bộ giỏ hàng
# Tạo menu cho người dùng nhập lựa chọn truy vấn
# từ 2 tới 7. Chương trình dừng khi người dùng
# nhập "yes" cho câu hỏi "Có muốn thoát khỏi
# chương trình không?
class ShoppingCart:
    gio_hang=[]
    def __init__(self):
        pass
    def them_vao(self,name,price):
        kq={"name":name,"price":price}
        ShoppingCart.gio_hang.append(kq)
        print(f"da them vao {kq["name"]}")
    def xoa_1_sp(self,name):
        if len(ShoppingCart.gio_hang)==0:
            return "khoong co san pham de xoa"
        else:
            for i in ShoppingCart.gio_hang:
                if i["name"]== name:
                    ShoppingCart.gio_hang.remove(i["name"])
                    break
    def kiem_tra_rong(self):
        return len(ShoppingCart.gio_hang)==0
    def tinh_tong_tien(self):
        s=0
        for i in ShoppingCart.gio_hang:
            s+=int(i["price"])
        return f"tổng tiền là : {s}"
    def hien_thi_sp(self):
        for i in ShoppingCart.gio_hang:
            print(f"sản phẩm:{i['name']} có giá:{i['price']}")
    def xoa_gio_hang(self):
        ShoppingCart.gio_hang.clear()
        return "gio hang da duoc xoa"
cart=ShoppingCart()
while True:
    n=input()
    if n=="1":
        print("ch co gi trong gio hang")
    elif n=="2":
        t=input()
        g=int(input())
        cart.them_vao(t,g)
    elif n=="3":
        t=input()
        cart.xoa_1_sp(t)
    elif n=="4":
        if cart.kiem_tra_rong():
            print("rong")
        else:
            print("kh rong")
    elif n=="5":
        print(cart.tinh_tong_tien())
    elif n=="6":
        cart.hien_thi_sp()
    elif n=="7":
        print('da xoa het gio hang')
    dungct=input()
    if dungct.lower()=="yes":
        break
```

```python
# Yêu cầu: Xây dựng lớp Rectangle(w, h) với các phương thức:
# area() trả về diện tích.
# perimeter() trả về chu vi.
# scale(k) phóng to hình chữ nhật theo hệ số k
# (nhân cả w và h), không trả về gì.
# Input
# w h k (số thực dương)
# Output
# area perimeter sau khi phóng to (làm tròn 2 chữ
# số thập phân).
class Rectangle:
    def __init__(self,a,b):
        self.w=a
        self.h=b
    def area(self):
        return f"dien tich truoc khi phong to la: {self.w * self.h}"
    def perimeter(self):
        return f"chu vi truoc khi phong to la: {(self.w+self.h)*2}"
    def scale(self,k):
        self.w=self.w*k
        self.h=self.h*k
        S2=self.w*self.h
        C2=(self.w+self.h)*2
        return (f"dien tich sau khi phong to la: {S2:.2f} \n"
                f"chu vi sau khi phong to la: {C2:.2f}")
w,h,k=map(float,input().split())
c=Rectangle(w,h)
print(c.area())
print(c.perimeter())
print(c.scale(k))
```

```python
# add_score(subject, score)
# avg() trả về điểm trung bình.
# rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu
# 6.5–<8, "Average" nếu 5–<6.5, ngược lại "Poor".
# Input
# Dòng 1: name
# Dòng 2: n
# n dòng tiếp: subject score (score là số thực 0–
# 10)
# Output
# name avg rank (avg 2 chữ số).
class student:
    def __init__(self,subject):
        self.a=[]
        self.s=subject
    def them_diem(self,score):
        self.a.append(float(score))
    def diem_tb(self):
        return sum(self.a)/len(self.a)
    def rank(self):
        if self.diem_tb()>=8:
            return "Excellent"
        elif self.diem_tb()<8 and self.diem_tb()>=6.5:
            return "good"
        elif self.diem_tb()<6.5 and self.diem_tb()>=5:
            return "Average"
        elif self.diem_tb()<5:
            return "poor"
    def __str__(self):
        return f"{self.s} {self.diem_tb():.2f} {self.rank()}"
a=input()
z=student(a)
n=int(input())
for i in range(n):
    subject,score=input().split()
    z.them_diem(float(score))
print(z)
```

```python
# [Triangle]
# Khai báo cấu trúc dữ liệu tam giác \(Triangle\)
# chứa \(3\) biến \(egde1,\,edge2,\,edge3\) kiểu \
# (int\) là độ dài \(3\) cạnh của tam giác đó.
# Viết hàm khởi tạo Triangle(int initEdge1,int
# initEdge2, int initEdge3) để khởi tạo độ dài các
# cạnh của tam giác và hàm double getArea() trả về
# diện tích của tam giác đó (nếu tam giác không
# hợp lệ, trả về \(-1\)).
# Hoàn thiện chương trình dưới đây nhập vào \(3\)
# cạnh của một tam giác và in ra diện tích của tam
# giác đó.
# Đầu vào
# Một dòng duy nhất từ bàn phím chứa \(3\) số
# nguyên là độ dài \(3\) cạnh của tam giác
# Đầu ra
# In ra màn hình diện tích tam giác (làm tròn đến\
# (2\) chữ số thập phân sau dấu phẩy). In ra \
# (invalid\) nếu \(3\) số nhập vào không phải là \(3\)
# cạnh của một tam giác
import math
class Triangle:
    def __init__(self,edge1,edge2,edge3):
        self.x=edge1
        self.y=edge2
        self.z=edge3
    def do_dai_hop_le(self):
        x1,x2,x3=self.x,self.y,self.z
        return x1>0 and x2>0 and x3>0 and (x1+x2>x3) and (x1+x3>x2) and (x2+x3>x1)
    def kiem_tra(self):
        if not self.do_dai_hop_le():
            print("-1")
    def double_getarea(self):
        p=(self.x+self.y+self.z)/2
        s=math.sqrt(p*(p-self.x)*(p-self.y)*(p-self.z))
        return s
    def __str__(self):
        if self.do_dai_hop_le():
            return f"{self.double_getarea():.2f}"
        else:
            return "invalid"
a,b,c=map(int,input().split())
z=Triangle(a,b,c)
print(z)
```

```python
# Yêu cầu
# Lớp BankAccount(owner, balance) với:
# deposit(x), withdraw(x); nếu rút quá số dư thì bỏ      deposit là nạp thêm tiền
# qua giao dịch.                                         withdraw là rút tiền
# __str__() in owner: balance.
# Input
# Dòng 1: owner initial_balance
# Dòng 2: n (số lệnh)
# Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc
# WITHDRAW x.
# Output
# Số dư cuối cùng (in số thực, 2 chữ số).
class BankAccount:                                  
    def __init__(self,owner,blance):                #owner là tên chủ tài khoản
        self.ten=owner                              #blance là số dư còn lại
        self.sodu=float(blance)
        self.a=[]
    def deposit(self,n):
        self.x=float(n)
        self.sodu+=self.x
        return float(self.sodu)
    def withdraw(self,e):
        if self.sodu>=float(e):
            self.sodu-=float(e)
        return self.sodu
ten,sodu=input().split()
z=BankAccount(ten,sodu)
n=int(input())
for i in range(n):
    e=input().split()
    if e[0].lower()=="deposit":
        z.deposit(e[1])
    elif e[0].lower()=="withdraw":
        z.withdraw(e[1])
print(f"{z.sodu:.2f}")

        
```
