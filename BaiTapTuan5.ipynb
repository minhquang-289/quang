```python
#W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
def check(n):
    for i in range(2,n-1):
        if n % i == 0:
            return False
        else:
            return True
if __name__=='__main__':
    n=int(input())
    print(check(n))
    
```

```python
#W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.
# SO hoan hao la so tong cac uoc bang chinh nó
def check(n):
    sum =0
    for i in range(1,n-1):
        if n % i ==0:
            sum=sum+i
    if sum ==n:
        return True
    else:
        return False
if __name__=='__main__':
    n = int(input())
    print(check(n))
```

```python
#W5A5:Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
#phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1
def lst(A,k):
    for i in range(len(A)):
        if A[i] == k:
            return i
    return -1
if __name__=='__main__':
    A =list(map(int,input().split()))
    k= int(input())
    print(lst(A,k))
```

```python
#W5A6:W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 
def gt(n):
    tich=1
    for i in range(1,n+1):
        tich=tich*i
    return tich
if __name__=='__main__':
    n=int(input())
    print(gt(n))
```

```python
#W5A7
#Viết hàm giả lập một máy tính bỏ túi. 
#Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử 
#\( operat \) 
#Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
#Phép trừ \( - \) 
#Phép cộng \( + \) 
#Phép chia \( / \) 
#Phép nhân \( * \) 
#Kết quả làm tròn đến chữ số thập phân thứ hai. 
#Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
#Đầu ra: Kết quả phép toán.
def mt(a,b,c):
    res =0
    if c == '+':
        res=a+b
    elif c == '-':
        res= a-b
    elif c == '/':
        if b == 0:
            print("Du lieu khong hop le!")
            return None
        else:
            res=a/b
    elif c == '*':
        res= a * b
    else:
        print("Loi toan tu khong hop le")
        return None
    return round(res,2)
if __name__=='__main__':
    a=float(input())
    b=float(input())
    c=input()
    mt(a,b,c)
    
    
```
